```python
from IPython.display import display_markdown
from xray_fov.paths import paths, outs
from xray_fov.data.dicom import dicom_directory
from xray_fov.plot.phantom_stats import plot_phantom_stats
```

```python
for key, path in paths["dicom"].items():
    dicoms = dicom_directory(path)
    display_markdown(f"# {key}", raw=True)
    fig = plot_phantom_stats(dicoms, noxray=("noxray" in key))
    display(fig)
    fig.savefig(outs["phantom-stats"]+f"/{key}.svg")
    fig.savefig(outs["phantom-stats"]+f"/{key}.jpg")
```

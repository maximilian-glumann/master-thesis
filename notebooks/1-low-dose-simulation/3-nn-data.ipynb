```python
from IPython.display import display_markdown
from xray_fov.paths import paths, outs
from xray_fov.data.image import image_directory
import matplotlib.pyplot as plt
```

```python
for key, path in paths["nn"].items():
    images = image_directory(path, preload=False, caching=False)
    for img in images["pixel_array"]:
        if(img.min() != 0 or img.max() != 255):
            print(img.min(), img.max())
```

```python
for key, path in paths["nn"].items():
    images = image_directory(path, preload=False, caching=False)
    display_markdown(f"# {key}", raw=True)
    display(images)
```

```python
for key, path in paths["nn"].items():
    images = image_directory(path, preload=False, caching=False)
    display_markdown(f"# {key}", raw=True)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7,3))
    if "noxray" in key:
        ax1.imshow(images[-1]["clamped_pixel_array"], cmap='gray')
        ax2.imshow(images[0]["clamped_pixel_array"], cmap='gray')
    else:
        ax1.imshow(images[0]["clamped_pixel_array"], cmap='gray')
        ax2.imshow(images[-1]["clamped_pixel_array"], cmap='gray')
    ax1.set_axis_off()
    ax2.set_axis_off()
    plt.tight_layout()
    plt.savefig(outs["nn-images"]+f"/{key}.svg")
    plt.savefig(outs["nn-images"]+f"/{key}.jpg")
    plt.show()
```
